```python
import pandas as pd
import numpy as np
```

```python
features = pd.read_csv("./train.csv") #Train data Loader 
print(features.shape) # Shape of the data
cols_features = features.columns # columns of the data 

labels = pd.read_csv("./trainLabels.csv") # Labels of the data
print(labels.shape) #shape of the labels of the data
```

```python
data = pd.merge(features,labels.iloc[:9999, :],on='id') #merging of the data
```

```python
data #print the data 
```

```python
data.head() #print head of the data 
```

```python
data.info() #information about the data 
```

```python
data = data.drop(['x3', 'x4', 'id'], axis=1) #dropping the data 
```

```python
data #priting the data after dropping unwanted columns
```

```python
from sklearn.preprocessing import LabelEncoder # lABEL ENCODER OF THE DATA 

lb = LabelEncoder() #intilizing the object of the data
```

```python
data = data.apply(lambda series: pd.Series(
    LabelEncoder().fit_transform(series[series.notnull()]),
    index=series[series.notnull()].index
)) #label encoder with out chaging into the numpy array this keep data as data frame
```

```python
data #printing the data after label encoding
```

```python
features_data = data.iloc[:, :143] # Separating all the features from the data 
```

```python
features_data #print features data 
```

```python
labels_data = data.iloc[:, 143:] #seaparting the labels from the data 
```

```python
labels_data #printing the labels of the data 
```

```python
nulls = list(data.isna().sum())  #checking for the null values of the data
cols = list(data.columns) #printing the columns of the data 
for i in range(len(cols)): #looping all the columns
    print(f"{cols[i]} and {data[cols[i]].dtype} : {nulls[i]/len(data[cols[i]]) * 100}") #printing the percentage of the null data in each column
```

```python
from sklearn.impute import KNNImputer #immputer of the data 
imputer = KNNImputer(n_neighbors=2) #creating the object for the imputer 
data = imputer.fit_transform(features_data) #fitting the data and transforming it into the null value filled data
```

```python
data #printing the sample of the data
```

```python
from sklearn.preprocessing import MinMaxScaler #importing minmaxscaler
sc = MinMaxScaler() #scaling the data
```

```python
data = sc.fit_transform(data) #transforming the data 
```

```python
print(f"Total number of NaN :- {np.isnan(data).sum()}") #checking for the null value for the final time
```

```python
featurestrain = data[:len(data)//2, :] #splitting the data into the train features
featurestest = data[len(data)//2:, :] #splitting the data into the test features


labelstrain = labels_data.iloc[:len(data)//2, :].values #splitting the data into the train labels 
labelstest = labels_data.iloc[len(data)//2:, :].values # splitting the data into the test labels 
```

```python
labelstrain.shape, featurestrain.shape #printing the shapes of the data
```

```python
def labels(data): #converting labels function
    y_train = [] #empty list
    for i in range(len(data)): #looping over the data 
        y_train.append(np.argmax(data[i])) #appending the label after converting from the categorical data normal data
    return y_train #returning the final data 

y_train = labels(labelstrain) #return for the train data of labels 
y_test = labels(labelstest) #return for the test data of labels 

trainla = labels(labels_data.iloc[:, :].values) #return for the data of labels 
cols_train = labels_data.columns #columns of the labels data   
```

```python
from sklearn.ensemble import VotingClassifier #voting classifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier #random forest, Adaboostclassifier
from sklearn.tree import DecisionTreeClassifier #Decision tree classifier 
from sklearn.svm import SVC #Support vector machine 
from sklearn.naive_bayes import GaussianNB #Guassian Bias

estimator = [] #Empty estimator Created
fr = RandomForestClassifier(n_estimators=500) #Created Object for the random forest classifier
estimator.append(('FR', 
                  RandomForestClassifier(n_estimators=500))) #appending the randomforest classifier into the empty classifier 
estimator.append(('SVC', SVC(C= 70, gamma ='auto', probability = True, max_iter=15000))) #appending support vector classifier into the empty classifier
estimator.append(('GB', GaussianNB())) #Appending gaussian naive bayes algorithm into the classifier 
estimator.append(('DTC', DecisionTreeClassifier())) #appending Decision tree classifier into the classifiers
estimator.append(('AB', AdaBoostClassifier(base_estimator=fr, n_estimators=400)))# Appending Adaboost classifier into the classifier 


Voting_classifier = VotingClassifier(estimator) #Creating object for the voting classifier
Voting_classifier.fit(data, trainla) #fitting the data into the classifier 
```

```python
y_pred = Voting_classifier.predict(featurestest) #prediciting the data with the model
```

```python
y_pred[:10] #predicting the some predicting the data 
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Test accuracy prepared by us :- ", accuracy_score(y_pred, y_test)) #accuracy score for the test data 
print("Test accuracy prepared by us :- ", accuracy_score(Voting_classifier.predict(data), trainla)) # accuracy for the train data
```

```python
print("************* Train classification report *****************")
print(classification_report(Voting_classifier.predict(data), trainla)) # classification report for the train data 
print()
print("************* Test classification report *****************")
print(classification_report(y_pred, y_test)) # classification report for the test data
```

```python
from keras.utils import to_categorical #to categorical
y_pred = to_categorical(y_pred, num_classes=33) #converting preding data to categorical
```
